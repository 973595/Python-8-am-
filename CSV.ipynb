```python
import csv 
csv_columns = ['Name','Age','Address']
dict = [
    {'Name':'Aayush','Age':23,'Address':'Kathmandu'},
    {'Name': 'Rambo','Age':33,'Address':'Kathmandu'}
]

csvfile = open("data.csv",'w')
writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
writer.writeheader()
for data in dict:
    writer.writerow(data)
csvfile.close()

```

```python
import csv 
data = []
with open('data.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        data.append(row)
data
```

```python
#concat
df1 = pd.DataFrame({'sn':[1,2,3],
                   'name':['Ram','Shyam','Hari'],
                    'add':['Kathmandu','Bhaktapur','Lalitpur']})
df2 = pd.DataFrame({'sn':[4,5,6],
                    'name':['Akash','Sabin','Harry'],
                     'add':['kathmandu','Bhaktapur','Lalitpur']})
df = pd.concat([df1,df2], axis = 0, ignore_index=True)
df
                    
```
